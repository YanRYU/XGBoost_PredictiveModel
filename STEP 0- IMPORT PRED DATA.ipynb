```
df1 = spark.read.format("csv").options(header = "true",#是否具有表頭，預設"是"
                                           inferSchema = "true",#是否需要自動辨識data type，預設"是"
                                           encoding= 'utf-8', #指定文檔編碼，預設"utf-8"
                                           delimiter = ',',#指定分隔符號，預設","
                                           nullValue= "null"
                                          ).load("dbfs:/FileStore/shared_uploads/XXX/XXX.csv")
display(df1)
```
